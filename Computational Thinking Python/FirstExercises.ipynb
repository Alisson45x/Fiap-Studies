```python
# First Exercise / Area of the circle

#Variables

radius = float(input("Radius of the circle: "))#5

#Calculating the area of ​​a circle

area = 3.14 * (radius ** 2)

#Output

print(f"The area of ​​the circle is {area:.2f}")
```

```python
# Second Exercise / Area of the rectangle

#Variables

side1= float(input("First side of rectangle: "))#10
side2= float(input("Seconde side of rectangle: "))#20

#Calculating the area of ​​a rectangle

area = side1 * side2

#Output

print(f"The Area of rectangle is {area:.2f}")
```

```python
#Thrid Exercise / Area of triangle 

#Variaveis
base = float(input("The base: ")) #10  
high = float(input("The high: ")) #10

#Calculating the area of ​​a triangle
area = (base * high) / 2

#Output
print(f"The Area of triangle is {area:.2f}")
```
