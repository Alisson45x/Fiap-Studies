```python
# 3 - 
grade = int(input("Digite a nota")) # 10
if grade >= 6:
    if grade >= 9:
        print("Foi aprovado com execelencia")
    else:
        print("Aprovado!")
else:
    print("Reprovado")
```

```python
# 4 -
numero = float(input("Digite um numero: "))
if numero >= 0:
    if numero % 2 == 0:
        print("Positivo e par")
    else:
        print("Positivo e impar")
else:
    if numero == 0:
        print("0(Zero) e par")
    else:
        print("Numero negativo")
```

```python
# 5 -
valor = int(input("Digite um valor: ")) # 450
if valor >= 200:
    vip = input("Voce é vip:[Sim/Não] ").lower() # sim
    if vip == "sim":
        print(f"Valor com desconto de 20%: {valor * 0.8}")
    else:
        print(f"Valor com desconto de 10%: {valor * 0.9}")
else:
    print("Valor nao precisa de desconto")
```
