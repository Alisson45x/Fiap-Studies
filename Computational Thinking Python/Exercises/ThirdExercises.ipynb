```python
# 4 -
numero = int(input("Digite um numero: "))
if numero >= 0:
    if numero / 2 % 0:
        print("Positivo e par")
    else:
        print("Positivo e impar")
elif numero == 0:
    print("0(Zero) e par")
else:
    print("Numero negativo")
```

```python
# 5 -
valor = int(input("Digite um valor: "))
if valor >= 200:
    vip = input("Voce é vip:[Sim/Não] ").lower()
    if vip == "sim":
        valorDesconto = valor * 0.8
        print(f"Valor com disconto de 20%{valorDesconto}")
    else:
        valorDesconto = valor * 0.9
        print(f"Valor com disconto de 10%{valorDesconto}")
else:
    print("Valor nao precisa de desconto")
```
