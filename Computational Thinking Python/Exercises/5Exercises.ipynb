```python
# ex 1
listaNumeros = []
for i in range(1, 6):
    numero = float(input("Digite o numero: "))
    listaNumeros.append(numero)
soma = 0
for k in listaNumeros:
    soma += k
print(f"A soma de todos os numeros é {soma}, e a media é {soma / 5}")
```

```python
# ex 2
par = []
impar = []
for i in range (1,11):
    numero = float(input("Digite o numero: "))
    if numero % 2 == 0:
        par.append(numero)
    else:
        impar.append(numero)
for k in par:
    print(k)
for k in impar:
    print(k)
```

```python
# ex 3 
listaAluno = []
listaSala = []
for i in range (1, 6):
    nome = input("Nome")
    idade = input("Idade")
    listaAluno.append(nome)
    listaAluno.append(idade)
    listaSala.append(listaAluno)
    listaAluno = []
print(listaSala)
```

```python
# ex 4
lista_Aluno = []
lista_Aluno_Sala = []
for i in range(1,4):
    nome = input("Nome: ")
    nota1 = float(input("Nota1: "))
    nota2 = float(input("Nota2: "))
    media = (nota1 + nota2) / 2
    lista_Aluno.append(nome)
    lista_Aluno.append(nota1)
    lista_Aluno.append(nota2)
    lista_Aluno.append(media)
    lista_Aluno_Sala.append(lista_Aluno)
    lista_Aluno = []
print(lista_Aluno_Sala)
```

```python
# ex 5
compras = []
compras_final = []
soma = 0
for i in range(1, 101):
    nome = input("Digite um nome e [Q] para sair: ")
    if nome.upper() == "Q":
        break
    preco = float(input("Digite o preço: "))
    soma += preco
    compras.append(nome)
    compras.append(preco)
    compras_final.append(compras)
for k in compras_final:
    print(k[0], " ",  k[1])
print(f"Valor final da compra {soma}")
    
```
