```python
#ex.4 pegar tempo e distançia e mostrar velocidade media em km/h

distanciaKm = float(input("Digite a distançia em km: ")) 
tempoHoras = float(input("Digite o tempo em horas: "))
velocidadeMedia = distanciaKm / tempoHoras
print(f'A velocidade média é {velocidadeMedia} Km/h')
```

```python
#ex.9
nome = input("Digite seu nome: ")
sobrenome = input("Digite seu sobrenome: ")
cpf = input("Digite seu cpf: ")
nomeValidar = input("Digite seu nome para validação: ")
sobrenomeValidar = input("Digite seu sobrenome para validação: ")
cpfValidar = input("Digite seu cpf para validação: ")
if nome == nomeValidar and sobrenome == sobrenomeValidar and cpf == cpfValidar:
    print("Acesso VIP permitido!")  
elif nome == nomeValidar and sobrenome == sobrenomeValidar:
    print("Cpf incorreto.")
elif nome == nomeValidar or sobrenome == sobrenomeValidar and cpf == cpfValidar:
    print("Nome ou sobrenome incorreto.")
```

```python
#ex.9.extra
nome = input("Digite seu nome: ")
sobrenome = input("Digite seu sobrenome: ")
cpf = input("Digite seu cpf: ")
while True:
    while True:
        nomeValidar = input("Digite seu nome para validação: ")
        if nomeValidar == nome:
            break
        else:
            print("Nome incorreto. Tente novamente.")
    while True:
        sobrenomeValidar = input("Digite seu sobrenome para validação: ")
        if sobrenomeValidar == sobrenome:
            break
        else:
            print("Sobrenome incorreto. Tente novamente.")
    while True:
        cpfValidar = input("Digite seu cpf para validação: ")
        if cpfValidar == cpf:
            break
        else:
            print("Cpf incorreto. Tente novamente.")
    if nome == nomeValidar and sobrenome == sobrenomeValidar and cpf == cpfValidar:
        print("Acesso VIP permitido!")  
        break
```
