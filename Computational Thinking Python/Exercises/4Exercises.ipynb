```python
# Ex 1: Calcule a soma dos numeros de 1 a 100.

soma = 0

for i in range(1, 101):
    soma += i
print(soma)
```

```python
# Ex 2 Crie um programa que imprima quias sao os numeros pare de 1 a 50 e calcule a soma dos valores pares.

soma = 0

for i in range(1, 51):
    if i % 2 == 0:
        print(i)
        soma += i
print(f"A soma de todos esses numeros {soma}")

    

```

```python
# Ex 3  escreva um programa que identifique quais valores sao impares em uma range 1 a 51 e cacule a media destes valores.

divisao = 0
soma = 0
for i in range(1, 52):
    if i % 2 == 1:
        divisao += 1
        soma += i
print(soma)
print(divisao)
print(f"A media dos numeros impares é igual a {soma / 26}")
```

```python
# Cp

numero1 = float(input("Digite o primeiro numero: "))
numero2 = float(input("Digite o segundo numero: "))
escolha = input("Escolha a operação\nADIÇÃO ( + )\nSUBTRAÇÃO ( - )\nMULTIPLICAÇÃO ( * )\nDIVISÃO ( / )").upper()

if escolha == "ADIÇÃO" or escolha == "+":
    print(f"A soma de {numero1} com {numero2} é igual a {numero1 + numero2}")
elif escolha == "SUBTRAÇÃO" or escolha == "-":
   print(f"A subtrção de {numero1} com {numero2} é igual a {numero1 - numero2}")
elif escolha == "MULTIPLICAÇÃO" or escolha == "*":
    print(f"A multiplicação de {numero1} com {numero2} é igual a {numero1 * numero2}")
elif escolha == "DIVISÃO" or escolha == "/":
    if numero2 == 0:
        print("ERRO!")
        print("Não é possivel dividir por 0")
    else:
        print(f"A divisão de {numero1} com {numero2} é igual a {numero1 / numero2}")
else:
    print("Opção invalida")
        
```
