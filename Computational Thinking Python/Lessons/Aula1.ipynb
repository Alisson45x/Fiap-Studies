```python
# Name of variable = input = Informaition text

# --Input variables--
# float = decimal values 
# int = integers
# str(string) = text

# --Output type--
# "print()" = Show something 

# --Condition--
# If xxxx:
# Elif xxx:
# else: 

# --Arithmetic operators--
# + = Addition, - = Subtraction, * = Multiplication, / = Division, % = Modulus, ** Exponentiation 

# --String manipulation--
# (:.xf) = Decides how many decimal places
```

```python
# --Input variables--
n1= 5
n2= 7
n3= 8

# Calculating the average
avarage = (n1 + n2 + n3) / 3

# Output
print(f"The avarage is {avarage:.2f}")
```

```python
# --Input variables--
n1 = float(input("Digite a priemira nota:"))
n2 = float(input("Digite a segunda nota:"))
n3 = float(input("Digite a terceira nota:"))

# Calculating the average
finalGrade= n1+n2+n3

# Output
print (f"First grade: {n1}")
print (f"Second grade: {n2}")
print (f"Third grade: {n3}")
print (f"The final grade is {finalGrade:.2f}")

# Condition
if finalGrade >= 7:
    print("Approved!")
else:
    print("Disapproved!")
```
